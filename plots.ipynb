```python
Tc = 2 / np.log(1+np.sqrt(2))
l = np.arange(0.1, Tc, 0.0001)
t = np.sinh(2/l)
t = 1 - np.power(t,-4)
t = np.power(t,1/8)
plt.plot(l, t, 'r--')
plt.xlabel("Temperature")
plt.ylabel("Magnetization",fontsize=14)
plt.show()
```

```python
plot = plt.figure(figsize=(12,4))
plt.plot(l, t, color='red', label = "exact solution")
plt.xlabel("Temperature")
plt.ylabel("|Magnetization|",fontsize=18)
plt.legend(loc='best',fontsize=12)
extraticks=[2.269]
plt.xticks(list(plt.xticks()[0]) + extraticks,rotation='vertical')
plt.ylim(ymax=1)
plt.ylim(ymin=-0.05)
plt.plot(np.arange(2.27, 3.27, 0.01),np.zeros(100),color='red')
plt.plot(2.269+np.zeros(35),np.arange(0., 0.35, 0.01),color='red')
```
